```python
# Dependencies and Setup
import pandas as pd
from sqlalchemy import create_engine
import psycopg2

# File to Load (Remember to Change These)
file_to_load = "data/IMDb ratings.csv"

# Read Purchasing File and store into Pandas data frame
ratings_df = pd.read_csv(file_to_load)

ratings_df.head()
```

```python
imdb_ratings_df = pd.DataFrame({"IMDb_id":ratings_df["imdb_title_id"],
                                   "Rating":ratings_df["mean_vote"]})
imdb_ratings_df.head()
```

```python
imdb_ratings_df.dropna(inplace = True)
```

```python
connection_string = "postgres:postgres@localhost:5432/customer_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
imdb_ratings_df.to_sql(name='IMDB_ratings', con=engine, if_exists='append', index=True)
```
