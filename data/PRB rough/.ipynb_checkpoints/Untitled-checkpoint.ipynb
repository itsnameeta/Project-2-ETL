```python
import pandas as pd
```

```python
df = pd.read_csv("golden_globe_awards.csv")
df1 = pd.read_csv("imdb.csv")
```

```python
df = df.dropna()
df1 = df1.dropna()
```

```python
df = df.loc[df["win"] != False]
```

```python
del df[ 'ceremony']
```

```python
del df['win']
```

```python
df.reset_index(drop=True, inplace=True)
df = df
df.to_csv('awards.csv')
# Data from golden_globe_awards is clean and stored in awards.csv file
```

```python
df1 = df1.loc[df1["language"] == "English"]
df1.reset_index(drop=True, inplace=True)
df1
```

```python
df1["film"] = df1["title"]
df2 = pd.merge(df, df1, on= "film")
# data from imbd is now clean and stored in df2
```

```python
df2.to_csv('merged.csv')
```
